GRADUATION YEAR OF STUDENT 

```python
import numpy as np
import pandas as pd
```

```python
datac = pd.read_csv("Final Lead Data.csv")
datac
```

```python
datac.info()
```

```python
datac.drop_duplicates(subset = 'Email', inplace = True)
datac
```

```python
datac.info()
```

```python
def convert_academic_year(year_str):
    year_mapping = {'1st year': 1, '2nd year': 2, '3rd year': 3, 'Final year': 4}
    return year_mapping.get(year_str, 0)  # Return 0 for unexpected values 

datac['What is your current academic year?'] = datac['What is your current academic year?'].apply(convert_academic_year)
datac['Academic Year'] = datac['What is your current academic year?'] + datac['Academic Year']
datac['Academic Year'].dtype
```

```python
date_format = '%m/%d/%Y %I:%M:%S %p'
datac['Created'] = pd.to_datetime(datac['Created'], format = date_format)
datac['Year_col'] = datac['Created'].dt.year
datac
```

```python
datac['Graduation year'] = 4 - datac['Academic Year'] + datac['Year_col']
datac
```

```python
datac.describe()
```

This shows that the maximum no. of students graduate in the year of 2024 then 2022, 2021 and the minimum of them in the year of 2019.
