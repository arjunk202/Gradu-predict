```python
Predictors = ['CGPA','Speaking Skills','ML Knowledge']
```

```python
new_data = pd.DataFrame(Data[Predictors])
new_data['Placement Status'] = Data['Placement Status']

X = new_data.drop('Placement Status', axis = 1) # so that only predictors are present
y = new_data['Placement Status']

X
```

```python
y
```

```python
new_data.isnull().sum()
```

```python
y = y.fillna(-1)
y.count()
```

```python
new_values = {
    'Placed': 1,
    'Not placed': 0,
}

y.replace(new_values, inplace=True)

y
```

```python
                                                                                    # since we have nulls in placement status col then 666 is 33.51% of 1987 so ~30% i will take as y_train
from sklearn.model_selection import train_test_split
np.random.seed(1)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 50) 
                                                                                    # bcos i want only to split the dataset into the train set so i just ignored the tests set that the train_test_split returns a a tuple with four elements: X_train, X_test, y_train, and y_test
```

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score
classifier = MultinomialNB()
classifier.fit(X_train, y_train)
```

```python
y_test
```

```python
y_predict_classifier = classifier.predict(X_test)
accuracy_classifier = accuracy_score(y_predict_classifier,y_test)
accuracy_classifier
```

```python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

c_svc = SVC()
c_svc.fit(X_train, y_train)
```

```python
y_predict2 = c_svc.predict(X_test)
```

```python
accuracy_svc = accuracy_score(y_predict2, y_test)
accuracy_svc
```

```python
from sklearn import tree
dt_classifier = tree.DecisionTreeClassifier()
dt_classifier.fit(X_train, y_train)
```

```python
y_predict3 = dt_classifier.predict(X_test)
```

```python
accuracy_dtclassifier = accuracy_score(y_test, y_predict3)
accuracy_dtclassifier
```

```python
from sklearn.ensemble import RandomForestClassifier
ra_classifier = RandomForestClassifier(n_estimators=45)
ra_classifier.fit(X_train, y_train)
```

```python
y_predict4 = ra_classifier.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_ra_classifier = accuracy_score(y_test, y_predict4)
accuracy_ra_classifier
```
